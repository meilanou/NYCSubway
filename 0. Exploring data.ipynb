## Hourly entries vs. exits

First I'm plotting the histograms of the frequencies of different entries/exits readings. So that I can see, among all the stations, what distribution the ridership has. From the histograms we can see both entries and exits are right-skewed (the mass of the distribution is concentrated on the left). And for both the medians are lower than the means. It can indicate that overall the planning for turnstiles installed did not match commuters' pattern. A normal (bell curve) distribuion is more desired as it indicateds the usage of turnstiles are more even. In <a href="1.%20Statistical%20test.ipynb" target="_blank">section 1</a> I will double check the normality with statistical test.

<i><u>I'm trying to make the code as reusable as possible. Below are 2 functions used to calculate and plot comparison histograms. In the future I'll omit explanation like this.</u></i>

```python
'''
Display count, total, median and mean for 2 side-by-side factored histograms.
'''
def print_stats(title0, array0, title1, array1):   
    print "{0}: Count = {1:,.0f}, Total = {2:,.0f}, Median = {3:,.0f}, Mean = {4:,.2f}" \
    .format(title0, len(array0), np.sum(array0), np.median(array0), np.mean(array0))
    print "{0}: Count = {1:,.0f}, Total = {2:,.0f}, Median = {3:,.0f}, Mean = {4:,.2f}" \
    .format(title1, len(array1), np.sum(array1), np.median(array1), np.mean(array1))
```

```python
'''
Plot 2 side-by-side factored histograms. I'm making them side-by-side 
for visual clearness.
'''
def plot_histograms(title0, array0, title1, array1):
    f, axes = plt.subplots(1, 2, figsize=(20, 7), sharex=True, sharey=True)
    
    max_data = np.r_[array0, array1].max()
    bin_size = int(np.log10(max_data))*20
    bins = np.linspace(0, max_data, bin_size)

    y_max_0 = np.max(np.bincount(array0))
    y_max_1 = np.max(np.bincount(array1))
    #y_lim_max = np.r_[y_max_0, y_max_1].max()*20

    axes[0].set(xlabel="{0}, Max: {1:,.0f}, Bin size: {2}".format(title0, np.max(array0), bin_size), \
                ylabel="Freq, Max: {0:,.0f}".format(y_max_0), \
                title="Histogram of {}".format(title0))
    sns.distplot(array0, kde=False, bins=bins, color=b, ax=axes[0])

    axes[1].set(xlabel="{0}, Max: {1:,.0f}, Bin size: {2}".format(title1, np.max(array1), bin_size), \
                ylabel="Freq, Max: {0:,.0f}".format(y_max_1), \
                title="Histogram of {}".format(title1))
    sns.distplot(array1, kde=False, bins=bins, color=r, ax=axes[1])

    plt.setp(axes, yticks=[])
    plt.tight_layout()
```

```python
ent_hr = list(df["ENTRIESn_hourly"])
ext_hr = list(df["EXITSn_hourly"])
print_stats('Hourly entries', ent_hr, 'Hourly exits', ext_hr)
plot_histograms('Hourly entries', ent_hr, 'Hourly exits', ext_hr)
```

## Ridership of May 2011

Throughout May 2011, did ridership vary by date? From the chart below we can tell it followed a repeated pattern that seemed aligned with day of week. I displayed the date of every Sunday. Looks like NYC Subway had lower ridership on Friday and Saturday. Notice that Memorial Sunday 05-30-11 presented an ultra low ridership. We may consider to use holiday as a candidate of prediction feature.. There were 10 raining days in this time frame, so I won't make too much sense to divide it into rain vs. no rain. 

```python
'''
plot the average entries of every 4-hour for a selected factor
'''
def plot_4_hour_average_entries_bar(x, y, factor, xticks=None, xticklabels=None):
    f, ax = plt.subplots()
    ax = sns.barplot(x=x, y=y)
    ax.set_xlabel(factor)
    ax.set_ylabel('Mean of 4-hour entries')
    ax.set_title('Average 4-hour entries of Each {}'.format(factor))
    if xticks <> None:
        ax.set_xticks(xticks)
    if xticklabels <> None:
        ax.set_xticklabels(xticklabels, rotation=35, ha='right')
```

```python
sundays_dates = df[df['day_week'] == 0]['DATEn'].unique()
sundays_count = len(sundays_dates)
first_sunday_index = (d.datetime.strptime(sundays_dates[0], '%m-%d-%y') - d.datetime(2011, 5, 1, 0, 0, 0)).days
sunday_indices = np.linspace(first_sunday_index, 7*(sundays_count-1)+first_sunday_index, sundays_count, dtype=np.float)
date_group = df.groupby(['DATEn'], as_index=False)['DATEn', 'ENTRIESn_hourly'].aggregate(np.mean)
plot_4_hour_average_entries_bar(date_group['DATEn'], date_group['ENTRIESn_hourly'], 'Date', xticks=sunday_indices, xticklabels=sundays_dates)
```

## Ridership of day of week

Let me group the ridership by week of day to see the repeated pattern. Below is the chart that shows the means of entries by day of week for May 2011. Yes, NYC Subway had lower ridership on Friday and Saturday. We may consider to use day of week as a candidate of prediction feature.

```python
day_week_group = df.groupby(['day_week'])[['day_week', 'ENTRIESn_hourly']].aggregate(np.mean)
week_days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
plot_4_hour_average_entries_bar(day_week_group['day_week'], day_week_group['ENTRIESn_hourly'], 'Day of Week', xticklabels=week_days)
```

## Average hourly entries - rain vs. no rain

How about ridership every 4 hours with the factor of rain or no rain? The chart below is showing the average ridership of every 4 hours when it rained or not. It seems when it rained, more people chose to take subway. But it doesn't look like the pattern varied too much when it rained or not.

```python
'''
plot the average entries of every 4-hour for 2 selected factors
'''
def plot_4_hour_average_entries_2bars(data, x, y, series, factor, kind):
    #f, ax = plt.subplots()
    g = sns.factorplot(x, y, series, data=data, kind=kind, size=6)
    g.set_axis_labels('Hour', 'Mean of 4-hour entries')
    titles = ['Average 4-hour entries by {}'.format(factor)]
    for ax, title in zip(g.axes.flat, titles):
        ax.set_title(title)
        ax.set_ylim(bottom=0)
```

```python
hour_rain_group = df[['hour', 'ENTRIESn_hourly', 'rain']]
hour_rain_group['Rained'] = np.where(hour_rain_group['rain'] == 1, '1 Yes', '2 No')
plot_4_hour_average_entries_2bars(hour_rain_group, 'hour', 'ENTRIESn_hourly', 'Rained', 'Rain', 'bar')
```

## Average hourly entries - weather conditions

Besides tain or not, another indicator could be the weather condition. Plotting ridership with this factor and the pattern varies more. This can be another candidate for prediction feature.

```python
hour_conds_group = df[['hour', 'ENTRIESn_hourly', 'conds']]
hour_rain_group['Rained?'] = np.where(hour_rain_group['rain'] == 1, '1 Yes', '2 No')
plot_4_hour_average_entries_2bars(hour_conds_group, 'hour', 'ENTRIESn_hourly', 'conds', 'Weather Condition', 'point')
```

## Ridership of each station

From the histograms above we know that the planning for turnstiles didn't match the real commute pattern. How about the planning for stations? Below is the chart of averages of 4-hour entries from the most used station to the least used station. Some stations are much busier, it's again right-skewed. So station may be a candidate for prediction feature.

```python
station_average = df.groupby(['station'], as_index=False)['ENTRIESn_hourly'].aggregate(np.mean).sort("ENTRIESn_hourly", ascending=False)
ax = plt.plot(station_average['ENTRIESn_hourly'])
plt.xlabel('Total of {} stations'.format(len(station_average)))
plt.ylabel('Mean of 4-hour entries')
plt.title('Average 4-hour entries of each station, from high to low')
```
