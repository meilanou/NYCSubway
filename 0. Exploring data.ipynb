###### First import libraries for mathematical operations and plotting, and initialize global variables.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.html.widgets import interact
import IPython
%matplotlib inline

#decimal_precision = 2
#np.set_printoptions(precision=decimal_precision)

sns.set(style="white", palette="colorblind")
b, g, r, p = sns.color_palette("colorblind", 4)

df = pd.read_csv('turnstile_weather_v2.csv', parse_dates=['datetime'])
#print df.describe()
```

## Basic stats

From to the queries below, we can see daily data is provided for May 2011. The turnstile counts were read every 4 hours for a total of 207 stations and 240 remote units. But interestingly, total count of entries almost double that of exits. Deeper investigation might be needed. But due to the scope of the project, I'm using entries data as major indicator of ridership. Comparing to average rainfall of New York, May 2011's rainfall was slightly lower. But its temperature stayed within normal range.

```python
date_unique = df['DATEn'].unique()
print 'Date range of data: {0} to {1}'.format(date_unique.min(), date_unique.max())
print 'Days of data:       {}'.format(len(date_unique))
print 'Sample hours:       {}'.format(list(df['hour'].unique()))
print 'Number of stations: {}'.format(len(df['station'].unique()))
print 'Number of remotes:  {}'.format(len(df['UNIT'].unique()))
print 'Total entries:      {:,.0f}'.format(df['ENTRIESn_hourly'].sum())
print 'Total exits:        {:,.0f}'.format(df['EXITSn_hourly'].sum())
print 'Rain days:          {}'.format(df.groupby(['DATEn'])['DATEn','rain'].aggregate(np.max)['rain'].sum())
print 'Mean precipitation: {:.5f} (inches)'.format(np.mean(df['meanprecipi']))
print 'Mean temperature:   {:.5f} (F)'.format(np.mean(df['meantempi']))
```

###### Method to display count, median and mean for side-by-side factored histograms.

```python
def print_stats(title0, array0, title1, array1):   
    print "{0} , Count: {1:,.0f}".format(title0, len(array0)), ", Total: {:,.0f}".format(np.sum(array0)), \
    ", Median: {:,.0f}".format(np.median(array0)), ", Mean: {:,.2f} ".format(np.mean(array0))
    print "{0} , Count: {1:,.0f}".format(title1, len(array1)), ", Total: {:,.0f}".format(np.sum(array1)), \
    ", Median: {:,.0f}".format(np.median(array1)), ", Mean: {:,.2f} ".format(np.mean(array1))
```

###### Method to plot 2 factored histograms. I'm making them side-by-side visual clearness.

```python
def plot_histograms(title0, array0, title1, array1):
    f, axes = plt.subplots(1, 2, figsize=(20, 7), sharex=True, sharey=True)
    
    max_data = np.r_[array0, array1].max()
    bin_size = np.linspace(0, max_data, max_data / 300)

    y_max_0 = np.max(np.bincount(array0))
    y_max_1 = np.max(np.bincount(array1))
    #y_lim_max = np.r_[y_max_0, y_max_1].max()*20

    axes[0].set(xlabel="{0}, Max: {1}".format(title0, np.max(array0)), \
                ylabel="Freq, Max: {0}".format(y_max_0))
    sns.distplot(array0, kde=False, bins=bin_size, color=b, ax=axes[0])

    axes[1].set(xlabel="{0}, Max: {1}".format(title1, np.max(array1)), \
                ylabel="Freq, Max: {0}".format(y_max_1))
    sns.distplot(array1, kde=False, bins=bin_size, color=r, ax=axes[1])

    plt.setp(axes, yticks=[])
    plt.tight_layout()
```

## Hourly Entries vs. Exits

Hourly entries/exits actually means the accumulated difference of 4 hours from the previous regular reading. Here I'm plotting the histograms to see, among all the stations, what kind of distribution the ridership has. Both histograms of entries and exits are right-skewed, the mass of the distribution is concentrated on the left. And both medians are lower than the means. It can indicate that the numbers of turnstiles installed were not well thought of.

```python
ent_hr = list(df["ENTRIESn_hourly"])
ext_hr = list(df["EXITSn_hourly"])
print_stats('Hourly Entries', ent_hr, 'Hourly Exits', ext_hr)
plot_histograms('Hourly Entries', ent_hr, 'Hourly Exits', ext_hr)
```

## Ridership of each station

If some turnstiles idled a lot, how should the ridership look like for all the stations? Below is the chart to display the distribution from the most used station to the least used station. Station can be a candidate for prediction feature.

```python
station_group = df.groupby(['station'], as_index=False)['ENTRIESn_hourly'].aggregate(np.mean).sort("ENTRIESn_hourly", ascending=False)
ax = plt.plot(station_group['ENTRIESn_hourly'])
plt.xlabel('{} stations'.format(len(station_group)))
plt.ylabel('Mean of 4-hour entries')
plt.title('Average entries of each station, from highest to lowest')
```

## Ridership of each date

Through out May 2011, how was the ridership distributed? From the chart below we can tell it followed a repeatale pattern that was aligned with day of week. I displayed the date of every Sunday. Looks like NYC Subway had higher ridership on Monday, Tuesday, Wednesday and Thursday. Notice that Memorial Sunday 05-30-11 presented an ultra low ridership.

```python
date_group = df.groupby(['DATEn'], as_index=False)['DATEn', 'ENTRIESn_hourly'].aggregate(np.mean)
ax = sns.barplot(x=date_group['DATEn'], y=date_group['ENTRIESn_hourly'])
ax.set_xlabel('Date')
ax.set_ylabel('Mean of 4-hour entries')
ax.set_title('Average entries of each date')
ax.set_xticks([1,8,15,22,29])
ax.set_xticklabels(['05-02-11', '05-09-11', '05-16-11', '05-23-11', '05-30-11'], rotation=35, ha='right')
```

## Ridership of weekdays and weekend

Let me group the ridership by week of day to see if it followed the same pattern. Below is the chart that shows the means of entries by day of week for May 2011. Yes...NYC Subway had higher ridership on Monday, Tuesday, Wednesday and Thursday. We can roughly say that day of week could be a candidate as prediction feature.

```python
day_week_group = df.groupby(['day_week'])[['day_week', 'ENTRIESn_hourly']].aggregate(np.mean)
ax = sns.barplot(x=day_week_group['day_week'], y=day_week_group['ENTRIESn_hourly'])
ax.set_xlabel('Day of week')
ax.set_ylabel('Mean of 4-hour entries')
ax.set_title('Average entries of day of week')
ax.set_xticklabels(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], rotation=35)
```

## Hourly ridership (every 4 hours)

Because turnstile counts were read every 4 hours, not really hourly counts. Instead of a 24-hour chart, I'm plotting a 4-hour bar chart to get a sense of how ridership varied by different time of the day. Below is the chart to show the means of entries of every 4 hours for May 2011. So we can see ridership was higher in the second half of the day. We can roughly say that hour of day could be a candidate as prediction feature.

```python
hour_group_mean = df.groupby(['hour'])[['hour', 'ENTRIESn_hourly']].aggregate(np.mean)
ax = sns.barplot(x=hour_group_mean['hour'], y=hour_group_mean['ENTRIESn_hourly'])
ax.set_xlabel('Hour of day')
ax.set_ylabel('Means of 4-hour entries')
ax.set_title('Average entries of every 4 hours')
```

## Hourly ridership (every 4 hours) - Rain vs. No Rain

How about ridership every 4 hours with the factor of rain or no rain? The chart below is showing the average ridership of every 4 hours when it rained or not. It seems when it rained, more people chose to take subway. But it doesn't look like the pattern varied too much when it rained or not.

```python
hour_rain_group = df[['hour', 'ENTRIESn_hourly', 'rain']]
hour_rain_group['Rained?'] = np.where(hour_rain_group['rain']==1, '1 Yes', '2 No')
fg = sns.factorplot('hour', 'ENTRIESn_hourly', 'Rained?', data=hour_rain_group, kind='bar')
fg.set_xlabels('Hour of day')
fg.set_ylabels('Means of 4-hour entries')
fg.set_titles('Average entries of every 4 hours, rained or not')
```

## Entries - Rain vs. No Rain

If taking "rain" into consideration, what kind of distribution the ridership would have? Here I'm plotting the histograms again. Both histograms of entries od raining and non-raining days are right-skewed, the mass of the distribution is concentrated on the left. And both medians are lower than the means. According to the numbers we can tell ridership was a little bit higher when it rained. But is it statistically significant? (to be continued in <a href="1.%20Statistical%20test.ipynb" target="_blank">section 1</a>)

```python
ent_hr_rain = list(df[df['rain'] == 1]['ENTRIESn_hourly'])
ent_hr_no_rain = list(df[df['rain'] == 0]['ENTRIESn_hourly'])
print_stats('Hourly Entries - RAIN', ent_hr_rain, 'Hourly Entries - NO RAIN', ent_hr_no_rain)
plot_histograms('Hourly Entries - RAIN', ent_hr_rain, 'Hourly Entries - NO RAIN', ent_hr_no_rain)
```

## Hourly ridership (every 4 hours) - weather conditions

Besides tain or not, another indicator could be the weather condition. Plotting ridership with this factor and the pattern varies more. This can be another candidate for prediction feature.

```python
hour_conds_group = df[['hour', 'ENTRIESn_hourly', 'conds']]
fg = sns.factorplot('hour', 'ENTRIESn_hourly', 'conds', data=hour_conds_group, kind='point', size=6, palette="BuPu")
fg.set_xlabels('Hour of day')
fg.set_ylabels('Means of 4-hour entries')
fg.set_titles('Average entries of every 4 hours by weather condition')
```
